```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam

n_features = X_train.shape[1]

# Model
model1 = Sequential()

# First Dense Layer
model1.add(Dense(100, activation='relu', input_shape=(n_features,)))
model1.add(BatchNormalization())
model1.add(Dropout(0.3))

# Second Dense Layer
model1.add(Dense(80, activation='relu'))
model1.add(BatchNormalization())
model1.add(Dropout(0.3))

# Third Dense Layer
model1.add(Dense(60, activation='relu'))
model1.add(BatchNormalization())
model1.add(Dropout(0.3))

# Fourth Dense Layer
model1.add(Dense(40, activation='relu'))
model1.add(BatchNormalization())
model1.add(Dropout(0.3))

# Fifth Dense Layer
model1.add(Dense(20, activation='relu'))
model1.add(BatchNormalization())
model1.add(Dropout(0.3))

# Output Layer
model1.add(Dense(5, activation='softmax'))
```

```python
model1.summary()
```

```python
# Early Stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1)
```

```python
# Compile Model
model1.compile(optimizer=Adam(learning_rate=0.001),
               loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
kv ={'Rear-end collision':0, 'Side impact collision':1, 'Run-off-road':2,
       'Overturn':3, 'Broadside collision':4}

y_train_int = y_train.map(kv)
```

```python
from tensorflow.keras.utils import to_categorical

y_train_encoded = to_categorical(y_train_int, num_classes=5)
```

```python

# Train Model
history = model1.fit(
    X_train, y_train_encoded,
    epochs=100, batch_size=64,
    validation_split=0.3,
    verbose=2,
    shuffle=True,
    callbacks=[early_stopping]
)
```

```python
y_test_int = y_test.map(kv)
```

```python
y_test_encoded = to_categorical(y_test_int, num_classes=5)
```

```python
# Evaluate Model
loss, acc = model1.evaluate(X_test, y_test_encoded, verbose=0)
print('Test Accuracy: %.3f' % acc)
```
